Answer to Qn3 unstacked % cohort analysis

1. Create a period column based on the email Date
Since we're doing quarterly cohorts, we'll be looking at the total quarterly behavior of our merchants. Therefore, we don't want granular OrderDate data (right now).

2. Determine the merchant's cohort group (based on their first order)
Create a new column called CohortGroup, which is the year and quarter in which the merchant's first purchase occurred.

3. Rollup data by CohortGroup & quarter
Since we're looking at quarterly cohorts, we need to aggregate merchants, orders, and amount spent by the CohortGroup within the quarter.

4. Label the CohortPeriod for each CohortGroup¶
We want to look at how each cohort has behaved in the quarters following their first purchase, so we'll need to index each cohort to their first purchase quarter. For example, CohortPeriod = 1 will be the cohort's first quarter, CohortPeriod = 2 is their second, and so on.

This allows us to compare cohorts across various stages of their lifetime.

5. User Retention by Cohort Group¶
We want to look at the percentage change of each CohortGroup over time -- not the absolute change.

-To do this, we'll first need to create a pandas Series containing each CohortGroup and its size.

-Now, we can utilize broadcasting to divide each column by the corresponding cohort_group_size.

-The resulting DataFrame, user_retention, contains the percentage of users from the cohort purchasing within the given period.

```python
def cohort_period(df):
    """
    Creates a `CohortPeriod` column, which is the Nth period based on the user's first purchase.
    
    Example
    -------
    Say you want to get the 3rd month for every user:
        df.sort(['UserId', 'OrderTime', inplace=True)
        df = df.groupby('UserId').apply(cohort_period)
        df[df.CohortPeriod == 3]
    """
    df['CohortPeriod'] = np.arange(len(df)) + 1
    return df
```

```python
appended_data2 = []
for Brand in buyers2:
    contain_values_2_new=contain_values_2_N.loc[contain_values_2_N['Brand']==Brand]
    contain_values_2_new.set_index('merchant_name', inplace=True)
    contain_values_2_new['CohortGroup'] = contain_values_2_new.groupby(level=0)['quarter'].min()
    contain_values_2_new.reset_index(inplace=True)
    grouped = contain_values_2_new.groupby(['Brand','CohortGroup', 'quarter'])
    # count the unique users, orders, and total revenue per Group + Period
    cohorts = grouped.agg({'merchant_name': pd.Series.nunique})
    # make the column names more meaningful
    cohorts.rename(columns={'merchant_name': 'Totalbuyers'}, inplace=True)
    cohorts = cohorts.groupby(level=1).apply(cohort_period)
    # reindex the DataFrame
    # reindex the DataFrame
    cohorts.reset_index(inplace=True)
    cohorts.set_index(['Brand','CohortGroup', 'CohortPeriod'], inplace=True)
    cohort_group_size = cohorts['Totalbuyers'].groupby(level=1).first()
    user_retention = cohorts['Totalbuyers'].unstack(1).divide(cohort_group_size, axis=1)
    appended_data2.append(user_retention)
appended_data2 = pd.concat(appended_data2)
```
